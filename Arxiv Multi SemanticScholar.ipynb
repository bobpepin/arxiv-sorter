```python
import re
from pathlib import Path
import xml.etree.ElementTree as ET
import sqltables.sqlite3
from tqdm.notebook import tqdm
```

```python
email_glob = "Accounts/**/Arxiv */*.xml"
```

```python
def parse_submission(submission):
    [_, meta, body, link] = re.split(r"\n\\\\", submission)
    meta = re.sub(r"\n  ", " ", meta)
    headers = [x.strip().split(":", 1) for x in re.findall(r"^\w+:.*?\n", meta, re.M)]
    headers = {k.strip(): v.strip() for k, v in headers}
    headers["arXiv"] = headers["arXiv"].replace(" (*cross-listing*)", "")
    body = body.strip()
    url = link.split(" ")[2]
    return headers, body, url
```

```python
db = sqltables.sqlite3.Database("submissions.sqlite3")
if "submissions" in list(db.tables):
    submissions = db.open_table("submissions")
else:
    submissions = db.create_table(name="submissions", column_names=["arxiv_id", "date", "title", "authors", "url", "abstract"])
```

```python
email_paths = list(Path(".").glob(email_glob))
for email_path in tqdm(email_paths):
    doc = ET.fromstring(email_path.read_text())
    data = doc.findtext(".//OPFMessageCopyBody")
    email_submissions = re.findall(r"\n\\\\\narXiv:.*?-{78}", data, re.DOTALL)
    email_rows = []
    for i, submission in enumerate(email_submissions):
        try:
            headers, body, url = parse_submission(submission)
            email_rows.append([headers["arXiv"], headers["Date"], headers["Title"], headers["Authors"], url, body])
        except Exception as e:
            print(f"Exception at {i}: {e!r}")
    submissions.insert(email_rows)
```
